## 1.      Create a list of 5 integers. Perform and print the result of the following operations: append, extend, insert, remove, pop, clear, index, count, sort, and reverse.

```python
nums = [10, 20, 30, 40, 50]
print("Original:", nums)

nums.append(60)          # add one element at end
print("After append:", nums)

nums.extend([70, 80])    # add multiple elements
print("After extend:", nums)

nums.insert(1, 15)       # insert at index 1
print("After insert:", nums)

nums.remove(30)          # remove first occurrence of value 30
print("After remove:", nums)

popped = nums.pop()      # remove last element
print("Popped value:", popped)
print("After pop:", nums)

# index of a value
print("Index of 40:", nums.index(40))   

# count of a value
print("Count of 20:", nums.count(20))   

nums.sort()              # sort ascending
print("After sort:", nums)

nums.reverse()           # reverse in-place
print("After reverse:", nums)

nums.clear()             # remove all elements
print("After clear:", nums)
```

### Section C: Exploring Pandas

```python
import pandas as pd
import numpy as np

data = {
    "Name":  ["Amit", "Neha", "Ravi", "Sara", "Amit", "Vijay", "Priya", "Ravi", "Kiran", "Neha"],
    "Age":   [25, 30, 22, 28, 25, 35, 27, 22, 29, 30],
    "City":  ["Delhi", "Mumbai", "Chennai", "Delhi", "Delhi", "Pune", "Kolkata", "Chennai", "Mumbai", "Mumbai"],
    "Salary": [50000, 60000, np.nan, 65000, 50000, 70000, 62000, np.nan, 58000, 60000]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

print("\n.info():")
print(df.info())

print("\n.describe():")
print(df.describe())  
```

```python
# Select 'Name' and 'City' columns
name_city = df[["Name", "City"]]
print("\nOnly Name and City columns:")
print(name_city) 

# Drop 'City' column (create a new DataFrame without City)
df_no_city = df.drop(columns=["City"])
print("\nDataFrame after dropping City column:")
print(df_no_city)  
```

```python
# Fill NaN in Salary with mean of Salary
salary_mean = df["Salary"].mean()
df["Salary"] = df["Salary"].fillna(salary_mean)

print("\nDataFrame after filling missing Salary with mean:")
print(df) 
```

```python
# Remove duplicate rows (keep first occurrence)
df_no_duplicates = df.drop_duplicates()

print("\nDataFrame after removing duplicate rows:")
print(df_no_duplicates) 
```

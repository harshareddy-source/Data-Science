### Exercise 1: Prime Numbers


```python
num = int(input("Enter a number: "))

if num <= 1:
    print(num, "is not a prime number")
else:
    is_prime = True
    i = 2
    while i * i <= num:
        if num % i == 0:
            is_prime = False
            break
        i += 1

    if is_prime:
        print(num, "is a prime number")
    else:
        print(num, "is not a prime number")
```

### Exercise 2: Product of Random Numbers

```python
import random

num1 = random.randint(1, 10)
num2 = random.randint(1, 10)

correct_product = num1 * num2

print("What is the product of", num1, "and", num2, "?")
user_answer = int(input("Your answer: "))

if user_answer == correct_product:
    print("Correct! Well done.")
else:
    print("Incorrect. The correct answer is", correct_product)
```

### Exercise 3: Squares of Even/Odd Numbers


```python

for num in range(100, 201):  # 201 so that 200 is included
    if num % 2 == 0:         # even number check
        square = num ** 2
        print(num, "squared is", square)
```

### Exercise 4: Hospital Billing System

```python

room_charge_per_day = 2000
doctor_fee = 1500
lab_test_charge_per_test = 300
discount_rate = 0.10  # 10%

# Patient details
days_stayed = 4
num_tests = 3
medicine_charges = 2400

# Calculate charges
room_charges = room_charge_per_day * days_stayed
lab_charges = lab_test_charge_per_test * num_tests

# Total before discount
total_before_discount = room_charges + doctor_fee + lab_charges + medicine_charges

# Apply discount if total before discount exceeds ₹10,000
if total_before_discount > 10000:
    discount_amount = total_before_discount * discount_rate
else:
    discount_amount = 0

final_bill = total_before_discount - discount_amount

print("\n----- BILL DETAILS -----")
print("Room charges        : ₹", room_charges)
print("Doctor consultation : ₹", doctor_fee)
print("Lab test charges    : ₹", lab_charges)
print("Medicine charges    : ₹", medicine_charges)
print("Total before discount: ₹", total_before_discount)
print("Discount applied    : ₹", discount_amount)
print("Final bill amount   : ₹", final_bill)
```

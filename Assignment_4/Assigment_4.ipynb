```python
import pandas as pd

# 1. Load datasets
patient_df = pd.read_csv("Patient_Data.csv")
billing_df = pd.read_csv("Billing_Data.csv")

# Show summary
patient_df.info()

# 2. Select billing-relevant columns
patient_billing = patient_df[['PatientID', 'Department', 'Doctor', 'BillAmount']]

# 3. Drop administrative columns
patient_no_admin = patient_df.drop(columns=['ReceptionistID', 'CheckInTime'])
```

```python
# 4. Total bill amount per department
bill_per_dept = patient_df.groupby('Department', as_index=False)['BillAmount'].sum()

# 5. Remove duplicate patients by PatientID
patient_unique = patient_df.drop_duplicates(subset='PatientID').copy()

# 6. Fill missing BillAmount with mean
mean_bill = patient_unique['BillAmount'].mean()
patient_unique['BillAmount'] = patient_unique['BillAmount'].fillna(mean_bill)
```

```python
merged = pd.merge(patient_unique, billing_df, on='PatientID', how='left')
```
